## Задание 3:
В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

### Вопрос 1

1. Дневная аудитория (DAU) — количество уникальных пользователей, которые зашли в приложение в течение суток во время тематического события. Характеризует количество активных игроков.
2. Stickness ratio — коэффициент того, насколько часто пользователи в течение тематического события пользователи заходили в игру.
3. Доля новых пользователей — отношение числа новых пользователей за время тематического события ко всем пользователям за этот же промежуток времени для отслеживания активности аудитории и восприятия ей ивентов в тематическом событии.
4. Retention rate (RR) — коэффициент удержания клиентов. На его основании можно проанализировать, например, возврат неактивных игроков, понять, какие ивенты им были интересны, постараться удержать этих пользователей.
5. Воронка — последовательность действий пользователей до совершения ключевого действия. От входа в игру до покупки предмета или подписки. Можно оценить конверсию игроков по событию.
6. ARPU - средний доход с привлечённого пользователя. Отслеживание финансововой ценности данного события.

### Вопрос 2

Я предполагаю, что данная механика негативно скажется на аудитории. Многим не понравится, что они при первой же неудачной попытке сразу вернуться на несколько уровней назад, тем самым, они возможно покинут игру после серии неудач. Однако возможно и другое поведение игроков, когда они, наоборот, после неудачи будут еще больше замотивированы на прохождение уровня.

Для понимания влияния данной механики на аудиторию необходимо к вышеперечисленным метрикам добавить следующие показатели:

1. Среднее время, проведенное пользователем в игре, чтобы понять, продолжают ли пользователи играть после неудачных попыток.
2. Воронка от стартовых до финальных уровней, чтобы отследить процент игроков, которые переходят между уровнями.
3. Retention rate для неудачных попыток, чтобы рассмотреть, какое количество пользователей удерживается после отката уровня.
